```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
#seperated the target column that is "chances of admission"
```

# Lets split the datas into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
X_train
```

```python
X_test
```

# Creating the model

```python
from sklearn import svm
#import the svm from scikit-learn
```

```python
clf = svm.SVR(gamma='scale', C=30, epsilon=0.0002,tol=0.001)
clf.fit(X_train, y_train) 
```

```python
import numpy as np
y_pred = clf.predict(X_test)
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
```

our MAE is 0.042

```python
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
```

our MSE is 0.004

```python
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

our RME is 0.063

```python
clf.score(X_test,y_test)
```

SCORE OF OUR MODEL IS 0.78
