```python
#lets drop the target column "quality"
X= data.drop("quality",axis=1)
```

```python
Y=data["quality"]
```

```python
#sp,lit the data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
#as the datas are dispersed, lets scale the datas using standardscaler()
from sklearn.preprocessing import StandardScaler
```

```python
scaler= StandardScaler()
```

```python
#scale the training data first
scaler.fit(X_train)
```

```python
scaler.mean_
```

```python
X_train_transformed = scaler.transform(X_train)
```

```python
#transformed training data is given below
```

```python
X_train_transformed
```

```python
#transform the test data also
X_test_transformed = scaler.transform(X_test)
```

```python
X_test_transformed
```

```python
#lets create the model using support vector classifier from scikit-learn
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(X_train_transformed,Y_train)
```

```python
#lets see the core
model.score(X_train_transformed,Y_train)
```

```python
Y_predict=model.predict(X_test_transformed)
```

```python
Y_predict
#the predicted values byb the model is below
```

```python
#lets see the accuracy
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,Y_predict)
```
